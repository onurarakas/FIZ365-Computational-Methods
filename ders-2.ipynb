```python
import numpy as np
import bilYonMod as bil
import matplotlib.pyplot as plt
```

```python
# Glocal Deikenler

mu = 3.98e5

def f(yVec, t):
    # yVec = [x, v_x, y, v_y]
    x = yVec[0]
    v_x = yVec[1]
    y = yVec[2]
    v_y = yVec[3]

    r = np.sqrt(x**2+y**2)

    return np.array([v_x, -mu*x/r**3, v_y, -mu*y/r**3])

#Balang koullar

t0 = 0

tson = 100000

y0Vec = np.array([-2500, 7.5, -5500, 0.5])

n = 100

#z

tTum, x_vx_y_vyTumVec = bil.add_coz_euler_sistem(f, t0, tson, y0Vec, n)

#yTumVec = [x(t), v_x(t), y(t), v_y(t)]

# izme

plt.plot(x_vx_y_vyTumVec[0], x_vx_y_vyTumVec[2], color = 'r', label = f'Euler adim sayisi:{n: .2e}')
plt.xlabel('x(t) [km]')
plt.ylabel('y(t) [km]')
plt.legend()
plt.grid()
```
